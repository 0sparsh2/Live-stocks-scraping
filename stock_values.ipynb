```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
stocks_url = 'https://economictimes.indiatimes.com/indices/nifty_50_companies'
stocks_response = requests.get(stocks_url)
stock_soup = BeautifulSoup(stocks_response.text.encode('utf-8'), 'html.parser')
```

```python
details = stock_soup.find_all('div', {'class': 'dataList'})
```

```python
#names = [detail.find('a').text for detail in details] 
#names
```

```python
x = stock_soup.find('div', {'class' : 'headers'})
```

```python
headings_column = [item.text for item in x.find('ul').find_all('li')]
len(headings_column)
```

```python
all_lists = []
for stock in details:
  lists = stock.find_all('li')
  all_lists.append([item.text for item in lists])

for list_vals in all_lists:
  list_vals[0] = list_vals[0][1:]
```

```python
nifty50 = pd.DataFrame(all_lists,columns=headings_column)
```
